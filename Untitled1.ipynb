```python
#drop unwanted colmunns
df = df.drop(['url', 'address', 'phone', 'menu_item', 'dish_liked', 'reviews_list'], axis = 1)
df.head()
```

## Droping Duplicates


```python
df.duplicated()
df.drop_duplicates(inplace = True)
df.shape
```

## Data Cleaning

```python
#cheakking for null values
df.isnull().sum()
```

```python
null_pctg = df.isnull().sum()/df.shape[0]*100
null_pctg
```

```python
#dealing with rate col
df['rate'].unique()
```

```python
#cleaning rate column
def handling_rate(value):
    if(value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate']  =  df['rate'].apply(handling_rate)
df['rate'].head()
```

```python
df['rate'].unique()
```

```python
#Filling Null Values in Rate Column with Mean

df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
#droping null values
df.dropna(inplace = True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns = {'approx_cost(for two people)':'Cost2plates', 'listed_in(type)':'Type'}, inplace = True)
df.head()
df.columns
```

```python
#deal with cost_for2 col

df['Cost2plates'].unique()
```

```python
def handling_cost(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
df['Cost2plates'] = df['Cost2plates'].apply(handling_cost)
df['Cost2plates'].unique()
```

```python
#Cleaning Rest Type Column

rest_types=df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handling_resttype(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handling_resttype)
df['rest_type'].value_counts()
```

```python
#Cleaning Location Column

Location =df['location'].value_counts(ascending = False)
Location

```

```python
location_lessthan300 = Location[Location<300]
location_lessthan300
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
        
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
#deal with cusine col
cuisines = df['cuisines'].value_counts(ascending  = False)
cuisines
```

```python
cuisines_lessthan100 = cuisines[cuisines<100]
cuisines_lessthan100
```

```python
def handling_cuisine(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handling_cuisine)
df['cuisines'].value_counts()
```
